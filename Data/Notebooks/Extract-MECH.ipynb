```python
import tabula
import pandas as pd
import string
```

```python
master=tabula.read_pdf("/content/bemechcurriculum.pdf", stream=True,pages='all')
```

```python
master[2]
```

```python
refined=[]
for num in range(8):
  if num==4:
    continue
  temp=master[num].drop(['Unnamed: 0','Unnamed: 1', 'Unnamed: 3','Unnamed: 4', 'Unnamed: 5', 'Unnamed: 7'], axis=1)
  if 'Unnamed: 8' in temp.columns:
    temp.drop(['Unnamed: 8'], inplace=True, axis=1)

  if(pd.isna(temp.iloc[3:4, -1]).bool()):
    temp.drop([0,1,2,3], axis=0, inplace=True)
  else:
    temp.drop([0,1,2], axis=0, inplace=True) 

  temp.columns=['Subject_Code','Subject','Credits']
  temp.dropna(axis=0, how="all", inplace=True)
  temp.reset_index(drop=True, inplace=True)  
  refined.append(temp[:-2])

len(refined)
```

**INDIVIDUAL CLEANING:**

```python
final=[]
```

```python
def nstrip(num):
  for val in range(len(refined[num])):
    refined[num]["Subject"][val]=refined[num]["Subject"][val].rstrip('0123456789')
```
